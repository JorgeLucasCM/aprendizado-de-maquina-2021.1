## Importando bibilhotecas



```python
import pandas as pd
import numpy as np
import random
import math
```

## Quest達o 1

```python
df = pd.read_csv('/content/Iris.csv')

setosa_len = len(df[df["Species"] == "Iris-setosa"])
versicolor_len = len(df[df["Species"] == "Iris-versicolor"])
virginica_len = len(df[df["Species"] == "Iris-virginica"])

print(setosa_len)
print(versicolor_len)
print(virginica_len)
```

## Quest達o 2

```python
treinoSet = df.sample(frac=0.8, random_state= 14)
testeSet = df.drop(treinoSet.index)

print(treinoSet)
print(testeSet)
```

## Quest達o 3



```python
treinoSet_mean = treinoSet.groupby(['Species']).mean()
treinoSet_mean
```

## Quest達o 4

```python
treinoMSL = treinoSet_mean['SepalLengthCm']
treinoMSW = treinoSet_mean['SepalWidthCm']
treinoMPL = treinoSet_mean['PetalLengthCm']
treinoMPW = treinoSet_mean['PetalWidthCm']

isSetosa = math.ceil(treinoMSL['Iris-setosa'] + treinoMSW['Iris-setosa'] + treinoMPL['Iris-setosa'] + treinoMPW['Iris-setosa'])

isVersicolor = math.ceil(treinoMSL['Iris-versicolor'] + treinoMSW['Iris-versicolor'] + treinoMPL['Iris-versicolor'] + treinoMPW['Iris-versicolor'])

isVirginica = math.ceil(treinoMSL['Iris-virginica'] + treinoMSW['Iris-virginica'] + treinoMPL['Iris-virginica'] + treinoMPW['Iris-virginica'])

testeSet['Species2'] = np.where( testeSet['SepalLengthCm'] + testeSet['SepalWidthCm'] + testeSet['PetalLengthCm'] + testeSet['PetalWidthCm'] <= isSetosa ,
                                'Iris-setosa',
                                np.where(testeSet['SepalLengthCm'] + testeSet['SepalWidthCm'] + testeSet['PetalLengthCm'] + testeSet['PetalWidthCm'] <= isVersicolor,
                                         'Iris-versicolor',
                                         'Iris-virginica'))

testeSet['Certo'] = testeSet['Species'] == testeSet['Species2']
accuracy = testeSet[testeSet['Certo'] == True]
precision = (len(accuracy)/len(testeSet['Certo'])) *100
print('Resposta com precis達o de',precision,'%')
```
